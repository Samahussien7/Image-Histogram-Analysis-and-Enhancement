```python

import matplotlib.pyplot as plt
import cv2

# Load the image
image = cv2.imread('/content/drive/MyDrive/cat.jpg')

# Convert the image to greyscale
greyscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Calculate the histogram
histogram = cv2.calcHist([greyscale_image], [0], None, [256], [0, 256])

# Plot the histogram
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("Bins")
plt.ylabel("# of Pixels")
plt.plot(histogram)
plt.xlim([0, 256])
plt.show()
```

```python

# Shift the histogram by a certain value
shifted_image = cv2.add(greyscale_image, 20)

# Calculate the histogram for the shifted image
histogram = cv2.calcHist([shifted_image], [0], None, [256], [0, 256])

# Plot the shifted histogram
plt.figure()
plt.title("Shifted Grayscale Histogram")
plt.xlabel("Bins")
plt.ylabel("# of Pixels")
plt.plot(histogram)
plt.xlim([0, 256])
plt.show()
```

```python


# Calculate the histogram
histogram = cv2.calcHist([greyscale_image], [0], None, [256], [0, 256])

# Calculate the cumulative histogram
cumulative_histogram = histogram.cumsum()

# Plot the cumulative histogram
plt.figure()
plt.title("Cumulative Grayscale Histogram")
plt.xlabel("Bins")
plt.ylabel("# of Pixels")
plt.plot(cumulative_histogram)
plt.xlim([0, 256])
plt.show()
```

```python


# Perform histogram equalization
equalized_image = cv2.equalizeHist(greyscale_image)

# Display the original and equalized images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(greyscale_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Equalized Image')
plt.imshow(equalized_image, cmap='gray')
plt.axis('off')

plt.show()
```

```python


# Calculate the histogram for the equalized image
equalized_histogram = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])

# Calculate the cumulative histogram for the equalized image
equalized_cumulative_histogram = equalized_histogram.cumsum()

# Plot the histogram and cumulative histogram for the equalized image
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.title('Equalized Image')
plt.imshow(equalized_image, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title('Equalized Image Histogram')
plt.xlabel('Bins')
plt.ylabel('# of Pixels')
plt.plot(equalized_histogram)
plt.xlim([0, 256])

plt.subplot(2, 2, 3)
plt.title('Equalized Image Cumulative Histogram')
plt.xlabel('Bins')
plt.ylabel('# of Pixels')
plt.plot(equalized_cumulative_histogram)
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```
